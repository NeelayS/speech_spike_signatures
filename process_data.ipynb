```python
class Dataset:
    
    def __init__(self, root_dir):
        self.root_dir = root_dir
        self.data = []
        
    def _frame(self, signal, n_frames=40, overlap=0.5):
        frame_length = int(len(signal)/(n_frames*(1-overlap) + overlap))
        frames = []
        start_index = 0 
        for i in range(n_frames):
            try:
                frame = signal[start_index:start_index+frame_length]

            except:
                frame = signal[start_index:]
                
            frames.append(frame)
            start_index += frame_length
         
        return frames
    
    def _featurize(self, signal):
        
    
    def preprocess(self):
        for file in os.listdir(self.root_dir):
            _, signal = wavfile.read(os.path.join(self.root_dir, file))
            self.data.append(self._frame(signal))
            
        return self.data
```
